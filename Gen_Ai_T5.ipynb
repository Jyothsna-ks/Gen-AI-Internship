```python
# Problem 1: Post Engagement Analyzer

def analyze_post_engagement(likes_list):
# This function calculates total likes and determines post is viral or normal

    total_likes = 0

    # Loop through likes list and calculate total
    for like in likes_list:
        total_likes += like

    # Check viral condition
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
   # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)
  # Sample Input
likes = [200, 300, 250, 300]
analyze_post_engagement(likes)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
# This function checks medicine stock level

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)
# Sample Input
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_checker(rainfall_list, required_level):
# This function calculates average rainfall and checks if it is adequate.
    total = 0

    # Calculate total rainfall
    for rain in rainfall_list:
        total += rain
    # Calculate average
    average = total / len(rainfall_list)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
# Sample Input
rainfall_data = [70, 75, 80, 65]
rainfall_checker(rainfall_data, 70)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicates(usernames):
# This function detects duplicate usernames

    unique_users = set(usernames)

    # If length differs, duplicates exist
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"
    print("Duplicate Accounts Found:", result)

# Sample Input
users = ["user1", "user2", "user1", "user3"]
detect_duplicates(users)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(price_list):
# This function filters crops above â‚¹2000

    premium_crops = []

    # Loop and filter
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)
# Sample Input
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def application_health_checker(error_count):
# This function checks system health based on error count.

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

# Sample Input
application_health_checker(7)
```

```python
# Problem 8: Daily Transaction Limit Checker

def transaction_checker(amount):
# This function validates daily transaction amount

    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Sample Input
transaction_checker(60000)
```

```python
# Problem 9: Attendance Eligibility System

def attendance_checker(attendance_list):
# This function calculates attendance percentage and checks exam eligibility.

    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100
    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Sample Input
attendance = ["P", "P", "A", "P", "P"]
attendance_checker(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
# This function calculates electricity bill

    bill = 0

    # First 100 units
    if units <= 100:
        bill = units * 3
    # Next 100 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    # Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)
# Sample Input
electricity_bill(250)
```
