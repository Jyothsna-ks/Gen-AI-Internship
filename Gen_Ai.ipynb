```python
# 1.User Login Check

# Stored credentials
username = "admin"
password = "1234"
# Input credentials
input_username = "admin"
input_password = "1234"
# Check login
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")

```

```python
# 2.Pass / Fail Analyzer

#Given marks of the students
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# 3.Simple Data Cleaner

#Input names
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()#Converts uppercase to lowercase
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
```

```python
# 4.Message Length Analyzer

#Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)#To check the length of the message
    print("Message:", msg)
    print("Length:", length)
if length > 10:
        print("Flag: Message longer than 10 characters")
```

```python

# 5.Error Message Detector

#Given error messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")#To count the no. of errors
print("Total ERROR entries:", error_count)
```
