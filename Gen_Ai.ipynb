```python
#TASK 2
#Problem 1: Unique Words in a Sentence

# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to get unique words
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding highest salary
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
total_days = len(attendance)
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

text = "python"
frequency = {}
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```

```python
# Problem 8: Convert List to Tuple

numbers = [10, 20, 30]
result = tuple(numbers)
print("Tuple:", result)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000
}
key_to_check = "Ravi"
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 75, 80, 65]
total = sum(marks)
average = total / len(marks)
print("Average Marks:", average)
```

```python
# 2.Pass / Fail Analyzer

#Given marks of the students
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# 3.Simple Data Cleaner

#Input names
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()#Converts uppercase to lowercase
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
```

```python
# 4.Message Length Analyzer

#Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)#To check the length of the message
    print("Message:", msg)
    print("Length:", length)
if length > 10:
        print("Flag: Message longer than 10 characters")
```
