```python
# Problem Statement 1: Employee Performance Bonus Eligibility
# Employee dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# To Find highest score
max_score = max(employees.values())
# To Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)
# Display result
print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {max_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis
from collections import Counter
import string

# Input query
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))
# Split into words
words = query.split()
# Count frequency using Counter
word_count = Counter(words)
# Filter words with frequency > 1
result = {word: count for word, count in word_count.items() if count > 1}
print(result)
```

```python
# Problem Statement 3: Sensor Data Validation
# Input list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

# Loop with index and value
for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:  # Check if even
        valid_readings.append((hour, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis
from collections import Counter

# Input list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Extract domains
domains = [email.split("@")[1] for email in emails]
# Count occurrences
domain_count = Counter(domains)
total_emails = len(emails)

# Display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Problem Statement 5: Sales Spike Detection
# Input list
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average = sum(sales) / len(sales)

# Step 2: Calculate threshold (30% above average)
threshold = average * 1.3

# Step 3: Detect spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem Statement 6: Duplicate User ID Detection
# Input list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count frequency
id_count = {}
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Display duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
