```python
# Problem 1: Smart Light Controller

# Class to control smart light
class SmartLight:

    def __init__(ownself, name):
        ownself.name = name     # light name
        ownself.status = "OFF"   # default status

    def turn_on(ownself):
        ownself.status = "ON"   # change to ON

    def turn_off(ownself):
        ownself.status = "OFF"  # change to OFF

    def show_status(ownself):
        # display current status
        print(ownself.name + " is " + ownself.status)

# Take Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")
light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()
light.show_status()
```

```python
# Problem 2: Employee ID Card System

# Class for storing the employee details
class Employee:

    def __init__(self_data, name, employee_id, department):
        # To store employee details
        self_data.name = name
        self_data.employee_id = employee_id
        self_data.department = department

    def display_id_card(self_data):
        # To print ID card details
        print("Employee ID Card")
        print("Name:", self_data.name)
        print("ID:", self_data.employee_id)
        print("Department:", self_data.department)

# Take Input
name = input("Employee Name: ")
employee_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, employee_id, department)
emp.display_id_card()
```

```python
# Problem 3: Mobile Contact Record

# Class to save contact details
class Contact:

    def __init__(ownself, name, phone_no):
        # To store contact details
        ownself.name = name
        ownself.phone_no = phone_no

    def save_contact(ownself):
        # To display saved contact
        print("Contact Saved")
        print("Name:", ownself.name)
        print("Phone:", ownself.phone_no)
# Take Input
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")
# Using Class
contact = Contact(contact_name, phone_number)
contact.save_contact()
```

```python
# Problem 4: Product Price Tag Generator

# Class to generate price tag
class Product:

    def __init__(own, name, price):
        # To store product details
        own.name = name
        own.price = price

    def print_price_tag(own):
        # To print formatted price
        print("Product:", own.name)
        print("Price: ₹" + str(own.price))
# Take Input
product_name = input("Product Name: ")
price = input("Price: ")
# Using Class
item = Product(product_name, price)
item.print_price_tag()
```

```python
# Problem 5: Movie Rating Display System

class Movie:

    def __init__(movie_s, name, rating):
        # To store movie details
        movie_s.name = name
        movie_s.rating = rating

    def display_movie(movie_s):
        # To show movie info
        print("Movie:", movie_s.name)
        print("Rating:", movie_s.rating, "/ 5")
# Take Input
movie_name = input("Movie Name: ")
rating = input("Rating: ")
# Use Class
movie = Movie(movie_name, rating)
movie.display_movie()
```

```python
# Problem 6: Delivery Address Manager
class Delivery:

    def __init__(data, customer_name, address):
        # To store delivery details
        data.customer_name = customer_name
        data.address = address

    def print_details(data):
        # To display delivery info
        print("Delivery Details")
        print("Customer:", data.customer_name)
        print("Address:", data.address)
# Take Input
customer_name = input("Customer Name: ")
address = input("Address: ")
# Using Class
delivery = Delivery(customer_name, address)
delivery.print_details()
```
