```python
# Problem Statement 1:Smart Parking Lot Management System

def smart_parking_system(capacity, vehicle_logs):
    # Function to manage parking lot entries and exits.
    # Parameters:: capacity (int): Maximum parking capacity
    # vehicle_logs (list): List containing "IN" and "OUT" logs
    parked = 0          # Current parked vehicles
    peak_usage = 0      # To track peak parking usage

    # Loop through each vehicle log
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
            # Update peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)
   # Check capacity status
    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)
# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def delivery_time_estimator(distance, traffic, weather):
# Function to estimate delivery time based on conditions

    base_time = distance * 5   # Assume 5 minutes per km

    # Add delay based on traffic
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10
    elif traffic.lower() == "low":
        base_time += 5

    # Add delay based on weather
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")
# Sample Input
delivery_time_estimator(6, "High", "Rainy")
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

def theatre_occupancy(total_seats, booked_seats):
# Function to calculate occupancy and show status.

    booked_count = len(booked_seats)   # Count booked seats
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

# Sample Input (150 seats booked out of 200)
booked_list = [1] * 150
theatre_occupancy(200, booked_list)
```

```python
# Problem Statement 4:Cloud Server Load Classification System
def server_load_classifier(cpu_readings):
# Function to classify server load based on average CPU usage.

   # Calculate average CPU load
    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classify load
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
# Sample Input
server_load_classifier([45, 60, 70, 85, 90])
```

```python
 # Problem Statement 5:Smart Classroom Resource Usage Monitor

def classroom_resource_monitor(resource_usage):
# Function to monitor classroom resource overuse

    overused = []  # List to store overused resources
    threshold = 8  # Define threshold

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
classroom_resource_monitor(usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
# Function to manage event registrations and prevent overbooking.

  # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration_controller(100, 105)
```
