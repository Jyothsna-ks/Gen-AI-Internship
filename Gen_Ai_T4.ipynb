```python
# Problem 1: Mobile Recharge Validation System
def recharge_system():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: "))

            if amount < 50:
                print("Recharge amount must be at least ₹50.")
            elif amount not in valid_plans:
                print("Invalid plan! Valid plans are 199, 299, 399, 599.")
            else:
                print(f"Recharge successful for ₹{amount}")
                break

        except ValueError:
            print("Invalid input! Please enter numbers only.")
recharge_system()
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")
# Sample dictionary
products = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 5,
    "Soap": 18
}

check_inventory(products)
```

```python
# Problem 3: Student Result Processing System
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"
marks = []
n = int(input("Enter number of subjects: "))
for i in range(n):
    m = int(input("Enter marks: "))
    marks.append(m)
avg, res = process_result(marks)

print("Average Marks:", avg)
print("Result:", res)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (per_km_rate * distance)
    if peak.lower() == "yes":
        fare += 0.25 * fare
    return fare

while True:
    try:
        distance = float(input("Enter distance (km): "))
        peak = input("Is it peak hour? (yes/no): ")
        total = calculate_fare(distance, peak)
        print("Total Fare: ₹", total)
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

    except ValueError:
        print("Please enter valid numeric distance.")
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
    present_count = 0
    for day in attendance_list:
        if day.upper() == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

attendance = []
n = int(input("Enter total number of working days: "))

for i in range(n):
    status = input("Enter attendance (P/A): ")
    attendance.append(status)
percent, result = check_eligibility(attendance)
print(f"Attendance Percentage: {percent}")
print("Status:", result)
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    digit_count = 0
    special_count = 0

    for char in password:
        if char.isdigit():
            digit_count += 1
        if char in "@#$":
            special_count += 1

    if len(password) < 8:
        return "Weak Password - Too Short"
    elif digit_count == 0:
        return "Weak Password - Add at least one digit"
    elif special_count == 0:
        return "Weak Password - Add at least one special character (@#$)"
    else:
        return "Strong Password"

password = input("Enter password: ")
print(check_password(password))
```
